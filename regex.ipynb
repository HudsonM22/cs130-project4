```python
import urllib.request
import re
import matplotlib.pyplot as plt
hand = urllib.request.urlopen('https://en.wikipedia.org/wiki/Neil_Armstrong')

def ascii_histogram(seq) -> None: #create a histogram from a list of numbers
  counted = count_elements(seq) #generate frequency table (dictionary)
  for key in sorted(counted):
    plus_str = '+' * counted[key]
    print(f'{key} {plus_str}') #prints out asterisk

def count_elements(seq): #creates a dictionary
    hist = dict()
    for date in seq:
      hist[date]=hist.get(date,0)+ 1
    return hist

dates = []
search_params = '(19\d{2}|20\d{2})'
for line in hand:
    line = line.decode().strip()
    dates.extend(re.findall(search_params, line))

dist1 = dates

# We can set the number of bins with the *bins* keyword argument.
n_bins = 80
fig, ax = plt.subplots(figsize=(10, 30))
ax.hist(dist1, bins=n_bins, orientation='horizontal')
ax.set(xlabel='counts',
       title='How many times each word appears')
plt.show()
```
